```python
#drops position and player
tm = nba.drop('POSITION','PLAYER')
```

```python
#Shows comparsion of rankVsalary between Orlando Magic
#Orlando Magic had the highest salary cap space and data shows they paid more for salaries for thier players
bn = tm.where('TEAM', are.equal_to('Orlando Magic')).sort('SALARY ($M)', descending=True)
bn.barh('RANK','SALARY ($M)')
# 5 players ranked below 100
```

```python
#Shows comparsion of rankVsalary between Los Angeles Lakers
#Lakers had max salary with Kobe Bryant so not much room for cap for other players.
lk = tm.where('TEAM', are.equal_to('Los Angeles Lakers')).sort('SALARY ($M)', descending=True)
lk.barh('RANK','SALARY ($M)')
#only 2 players ranked below 100
```
